```python
import pandas as pd
import numpy as np
import plotly.express as px
import os

# Carrega dados com caminho robusto
csv_path = os.path.join(os.path.dirname(os.getcwd()), 'vehicles.csv')
df = pd.read_csv(csv_path)

df.head()
```

```python
df.info()
df.describe()
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df)) * 100

pd.DataFrame({
    'missing': missing,
    'pct': missing_pct
})
```

```python
df = df.dropna(subset=['price', 'odometer'])
```

```python
fig = px.histogram(df, x='price', nbins=50, title='Distribuição de Preço')
fig.show()
```

```python
fig = px.histogram(df, x='odometer', nbins=50, title='Distribuição de Odometer')
fig.show()
```

```python
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1

upper = q3 + 1.5 * iqr

df_filtered = df[df['price'] <= upper]
```

```python
fig = px.scatter(
    df_filtered,
    x='odometer',
    y='price',
    title='Preço vs Quilometragem'
)
fig.show()
```

```python
df['condition'].value_counts()
```

```python
fig = px.box(
    df_filtered,
    x='condition',
    y='price',
    title='Preço por condição'
)
fig.show()
```
